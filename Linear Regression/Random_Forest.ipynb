```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dataset = pd.read_csv("Position_Salaries.csv")
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:,2].values
```

```python
from sklearn.ensemble import RandomForestRegressor
random_reg = RandomForestRegressor(n_estimators=10, random_state=0)
```

```python
random_reg.fit(X,y)
```

```python
y_pred = random_reg.predict(X)
y_pred
```

```python
plt.scatter(X,y, color = 'red')
plt.plot(X,random_reg.predict(X),color="blue")
plt.title("Truth or Bluff(Random Forest Regressor)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
```
